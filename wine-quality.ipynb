```python
bins = (2,6,8)
group_name = ['bad', 'good']
wine['quality']=pd.cut(wine['quality'], bins=bins, labels=group_name)

```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
wine_quality = wine['quality']
```

```python
label_quality = LabelEncoder()
```

```python
#tranform the string labels bad, average and good to 0,1,2
wine['quality'] = label_quality.fit_transform(wine['quality'])
```

```python
wine['quality'].value_counts()
```

```python
sns.countplot(wine['quality'])
```

```python
X = wine.drop('quality', axis = 1)
y = wine['quality']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.25)
```

```python
scaler = StandardScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, multilabel_confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier
```

### Random Forest Classifier 

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
print(confusion_matrix(y_test,pred_rfc))
```

```python
print(multilabel_confusion_matrix(y_test,pred_rfc))
```

### Gradient Descent

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd = SGDClassifier(penalty = None)
sgd.fit(X_train, y_train)
pred_sgd = sgd.predict(X_test)
```

```python
print(classification_report(y_test, pred_sgd))
```

### Support Vector Classifier

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
pred_svc = svc.predict(X_test)
```

#### Lets try to improve our accuracy of the model

```python
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel': ['linear', 'rbf'],
    'gamma': [0.1, 0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc, param_grid = param, scoring = 'accuracy', cv = 10)
```

```python
grid_svc.fit(X_train, y_train)
```

```python
grid_svc.best_params_
```

```python
svc2 = SVC(C = 1.2, gamma = 0.9, kernel = 'rbf')
svc2.fit(X_train, y_train)
pred_svc2 = svc2.predict(X_test)
print(classification_report(y_test, pred_svc2))
```

SVC improves fom 86% to 90% using GridCV

```python
rfc_eval = cross_val_score(estimator = rfc, X=X_train, y=y_train, cv=10)
rfc_eval.mean()
```

Random forest accuracy increases from 87% to 91 % using cross validation score
